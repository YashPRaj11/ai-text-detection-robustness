```python
human_df = pd.read_csv("../data/raw/human_text.csv")
ai_df = pd.read_csv('../data/raw/ai_text.csv')
```

```python
human_df.head()
```

```python
ai_df.head()
```

```python
## Cleaning both the dataset human_text and ai_text
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
human_df['Text'] = human_df['Text'].apply(clean_text)
ai_df['Text'] = ai_df['Text'].apply(clean_text)
```
