```python
import pandas as pd

human_df = pd.read_csv("../data/processed/human_text_cleaned.csv")
ai_df = pd.read_csv("../data/processed/ai_text_cleaned.csv")
```

```python
human_df['label'] = 0
ai_df['label'] = 1
```

```python
final_df = pd.concat([human_df, ai_df], ignore_index=True)
```

```python
## Shuffle index so human and Ai sample mixed randomly
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
## 70% train, 15% validation, 15% test
train_df, temp_df = train_test_split(final_df,test_size=0.3,random_state=42,stratify=final_df["label"])
```

```python
val_df, test_df = train_test_split(temp_df,test_size=0.5,random_state=42,stratify=temp_df["label"])
```

```python
final_df.to_csv("../data/processed/final_dataset.csv", index=False)
train_df.to_csv("../data/processed/train.csv", index=False)
val_df.to_csv("../data/processed/val.csv", index=False)
test_df.to_csv("../data/processed/test.csv", index=False)
```
